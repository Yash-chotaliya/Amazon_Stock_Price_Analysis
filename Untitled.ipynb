```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./dataset/data.csv')
df
```

```python
df.info()
```

```python
df['Day'] = df['Date'].apply(lambda x:x.split('-')[2])
df['Month'] = df['Date'].apply(lambda x:x.split('-')[1])
df['Year'] = df['Date'].apply(lambda x:x.split('-')[0])
```

```python
df['Day'] = df['Day'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)

df = df[df['Year']>1997]
df
```

```python
grp = df.groupby(['Year','Month'])
```

```python
# create new data frame which contains average open stock price for year 1998-2022

dict = {'Year':10,
        'Month':10,
        'Mean':10
       }
  
df1 = pd.DataFrame(dict,index=[0])
for i in range(2023-1998):
    for j in range(12):
        mean = grp.get_group((i+1998,j+1))['Open'].mean()
        df1.loc[len(df1.index)] = [int(i+1998),int(j+1),mean]
```

```python
df1.drop(0,axis=0,inplace=True)
```

```python
df1['Year'] = df1['Year'].astype(int)
df1['Month'] = df1['Month'].astype(int)
df1
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df1,x='Year',y='Mean',hue='Month')
```

```python
df1.to_csv('./dataset/year_wise_stock.csv',index=False)
```
